# Recipe Website Traffic Prediction

## Context
A (fictional) recipe website called "Tasty Bytes" features 1 recipe each day on its homepage. The featured recipe's Product Manager discovered that certain popular recipes being featured on the homepage can cause as much as a 40% increase in traffic to the rest of the website. 

Given previous data on the featured recipes, develop a model to predict which recipes will be ''popular''; that is, which recipes will lead to an increase in traffic. The product manager has expressed that they would like these predictions to be correct at least 80% of the time. 

## Data
All data from the previously featured recipes can be found in the client-provided file `recipe_site_traffic_2212.csv`. It contains 947 rows, with the following columns:

|Column       |Data Type        |Description   |
|-------------|-----------------|--------------|
|`recipe`     |Numeric (integer)| Unique identifier for the recipe. |
|`calories`   |Numeric (float)  | Number of calories in the recipe. |
|`carbohydrate`|Numeric (float) | Amount of carbohydrates (grams) in the recipe.|
|`sugar`      |Numeric (float)  | Amount of sugar (grams) in the recipe.|
|`protein`    |Numeric (float)  | Amount of protein (grams) in the recipe.|
|`category`   |Object (string)  | Recipe category. Ten possible values: Lunch/Snacks, Breakfast, Chicken, Meat, Dessert, Vegetable, Pork, Potato, Beverages, One Dish Meal|
|`servings`   |Numeric (integer)| Number of servings in the recipe.|
|`high_traffic`|Object (string) | Target variable. A value of 'High' indicates high traffic; left blank otherwise.|

## Procedure
To create this report, we will be adhering to the following procedure, informed by industry best practices and sound statistical theory:

1. Execute standard data validation and cleaning to prepare the data for analysis.
2. Use exploratory data analysis techniques to gain expertise over the data domain and observe possible patterns in the data.
3. Execute data preprocessing and model development according to best practices in data science; create both a baseline and comparison model.
4. Choose an appropriate model metric and evaluate the performance of the models.
5. Translate the cleaned data and model performance into business-ready insights:
    - Define appropriate KPI(s) to relate the models' performances to how they can increase revenue for Tasty Bytes.
    - Describe how the models perform with this approach. 
    - Furnish a final summary, including recommendations for Tasty Bytes to undertake. 

```python
# Import necessary libraries for data analysis, cleaning, ML
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import GridSearchCV

#
import warnings
warnings.filterwarnings("ignore")

# Read in the data
df = pd.read_csv('../DATA/recipe_site_traffic_2212.csv')
```

# I. Data validation
After reading in the data, we note that there are a total of 947 rows (recipes). Of those, we note that the columns `calories`, `carbohydrate`, `sugar`, `protein`, and `high_traffic` contain missing values. Using the `info` method of the obtained DataFrame, we can see that the data types are in line with expectations of the provided data and that there are no additional "stray" markers being used to signify missing values (e.g., using a dash or period).

A preliminary inspection of the DataFrame highlights two groups of columns where there are missing data points for different reasons: the numeric columns `calories`, `carbohydrate`, `sugar`, and `protein` contain true missing values, whereas the column `high_traffic` is using null values to provide additional information. 

The other type of data error present is values that are outside the expected range of values and data types, and these instances take a similar approach to fix. 

```python
# Initial inspection
print(df.info())
print(df.describe())
print('Distinct values in category column:', df['category'].unique())
print('Distinct values in high_traffic column:', df['high_traffic'].unique())
print('Distinct values in servings column:', df['servings'].unique())
print('Number of distinct values in servings column:', len(df['recipe'].unique()))
```

## I.a. The `recipe` column

There are 947 distinct integer entries in this column, as expected. No action needed. 

## I.b. The nutrition information columns

The numeric columns regarding nutritional information are either all missing at once or are all present. The nutritional information is missing in 52 instances out of 947, or around 5.49% of the data. In addition, it does not seem as though a pattern is present with respect to *which* recipes are missing data. Therefore, we elect to drop the data with missing nutritional information. 

```python
# Drop the rows with missing nutritional information
df_clean = df.loc[~df['calories'].isna(), :].reset_index()
print(df_clean.info())
```

## I.c. The `category` column

One item of note is that the `category` column is supposed to take on one of ten possible values: `'Lunch/Snacks'`, `'Beverages'`, `'Potato'`, `'Vegetable'`, `'Meat'`, `'Chicken'`, `'Pork'`, `'Dessert'`, `'Breakfast'`, and `'One Dish Meal'`. In the imported data, however, there is one additional value, `'Chicken Breast'`. It is clear that this is not intended to be a separate category, so we can safely change any occurrences of `'Chicken Breast'` to `'Chicken'`.

```python
# Fix erroneous extra values
df_clean['category'] = df['category'].str.replace('Chicken Breast', 'Chicken')
```

## I.d. The `servings` column

The initial inspection reveals that the column containing the number of servings has the `object` data type; this value is supposed to be an integer. Listing the unique values in this column, we can see that the issue is that there are two values not fitting the convention of the column and have the string "as a snack" appended after the number of servings. In this instance, we can simply strip that trailing string and re-cast the column as an integer type, as is intended.

```python
# Fix inconsistent serving entries
df_clean['servings'] = df['servings'].str.strip(' as a snack').astype('int')
```

## I.e. The `high_traffic` column

The column `high_traffic` contains a high number of missing values, but the `unique` method of the column series reveals that there are precisely two values: either `'High'` or `Null`. An entry was either recorded as `'High'` or left blank; this means that `high_traffic` is functioning more like a Boolean indicator, so to process this column, we will cast the column as an `int` type by replacing `'High'` with 1 and `Null` with 0. 

```python
# Encode 'high_traffic' column
df_clean['high_traffic'] = pd.get_dummies(df['high_traffic']).astype('int')

# Another inspection
print(df_clean.info())
```

# II. Exploratory data analysis

## II.a. Representation of recipe categories in the data

Although there are 10 different categories, it is important to know how well-represented each of the individual categories is.

As we can see in the plot below, the count across categories is fairly consistent, with the exception that there are many more `'Chicken'` dishes present in the data than recipes of any other category. 

```python
sns.set_style('darkgrid')
sns.set_palette('viridis')
sns.displot(data = df_clean, y = 'category')
plt.xlabel('')
plt.ylabel('Recipe Category')
plt.title('Counts of Recipes by Category')
plt.show()
```

When we account for the `'high_traffic'` rating within categories, we can get a better sense of the extent to which recipe category can affect popularity. In the chart below, we see the count plot from above but with an additional plot of the count of those recipes that were popular. This does gives us some additional information: for example, almost all of the `'Potato'`, `'Pork'`, and `'Vegetable'` recipes had a positive `'high_traffic'` indication. The `'Chicken'` category recipes had a similar number of recipes with positive `'high_traffic'` indication, but this is because this particular category had the most observations in the data set; the category of `'Chicken'` recipes had a `'high_traffic'` rate of less than 50%.

There are also potentially some relationships between the `'calories'`, `'carbohydrate'`, `'sugar'`, and `'protein'` features, but their pairwise Pearson correlation coefficients are near 0, which means that the relationship (if it exists) is unlikely to be linear.

# III. Model development

The goal is to predict whether a recipe featured on the front page is likely to be a high-traffic or "popular" recipe. Since the target variable is binary in nature (high-traffic vs. not high-traffic), the machine learning problem at hand is a **classification** problem.

Before fitting any models, we will do some preprocessing of the data, namely:
- Encode the `'category'` into integer values using `LabelEncoder` from `sklearn.preprocessing`.
- Declare the DataFrame of features `X` and target `y`. 
- Split the data into training and testing data using `train_test_split` from `sklearn.preprocessing`.
- Fit a `StandardScaler` to the training data, and *then* transform the training and test data to avoid data leakage.

```python
# Get categorical dummy variables
encode = LabelEncoder()
df_clean['category'] = encode.fit_transform(df_clean['category'])

features = ['calories', 'carbohydrate', 'sugar', 'protein', 'category', 'servings']

X = df_clean[features]
y = df_clean['high_traffic']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=6122)

scale = StandardScaler()

numeric_cols = ['calories', 'carbohydrate', 'protein', 'sugar']

# Fit & transform training data
X_train[numeric_cols] = scale.fit_transform(X_train[numeric_cols])

# Transform only on test data
X_test[numeric_cols] = scale.transform(X_test[numeric_cols])
```

## III.a. Baseline Model: Logistic Regression
As a baseline classifier, we will fit a basic `LogisticRegression` estimator, which is an estimator that is easy to explain and interpret. This returns a precision of about 83%, which is within the desired range. 

```python
# Initialize the classifier
logit = LogisticRegression()

# Fit to the training data
logit.fit(X_train, y_train)

y_pred = logit.predict(X_test)

# Get precision score
prec = precision_score(y_test, y_pred)
print(prec)
```

## III.b. Comparison Model: Decision Tree Classifier

As a comparison model, we will be using a decision tree classifier. These classifiers are computationally quick and although a bit less straightforward than logistic regression, are still fairly straightforward to understand. Decision trees are prone to overfitting, so a `GridSearchCV` was used during hyperparameter tuning (see Section IV.b.). This model returned a precision of about 82%, which is slightly less than the logistic regression model, but still within the desired range. 

```python
# Initialize the decision tree
dt = DecisionTreeClassifier(max_depth=3, min_samples_leaf=3, random_state=6122)

dt.fit(X_train, y_train)

y_pred_rf = dt.predict(X_test)
prec_dt = precision_score(y_test, y_pred_rf)
print(prec_dt)
```

# IV. Model Evaluation
The metric we are using to gauge model performance is **precision**, which is the ratio of 'true positives' and 'total positives' -- in our case, we get a percentage of recipes predicted to be popular (high-traffic) that actually were high-traffic values. 

This metric is more valuable than accuracy in the client's case. This is because the current practice is to choose a single recipe to feature on the homepage. It's therefore not possible to choose all popular recipes on a given day, and so there will always be popular recipes left out of selection. In particular, a false negative result (predicting a recipe would not be popular when it actually is) does not actively hamper an increase in website traffic.

By contrast, a false positive (predicting a recipe would be popular and then it is not) has a detrimental effect on potential revenue, because in this scenario an uptick in web traffic has been prevented by an incorrect prediction. The client's aim was to "correctly predict high traffic recipes 80% of the time". Thus, the highest priority should be in maximizing the ratio of true positives and total positives, i.e., the model's precision. 

```python
logit_coeffs = logit.coef_[0]

results_dict = {}

# Collect the coefficients in a dictionary for plotting
for i in range(len(features)):
    results_dict[features[i]] = float(logit_coeffs[i])
    
# Convert dictionary keys and values to lists
keys = list(results_dict.keys())
values = list(results_dict.values())

# Plot the logistic regression coefficients
sns.barplot(x=keys, y=values)
plt.title('Feature importance in logistic regression')
plt.show()
```

In contrast to the logistic regression above, one feature in the decision tree model heavily outweighs the others: the recipe `category`. 

```python
dt_feat = {}
for i in range(len(features)):
    dt_feat[features[i]] = dt.feature_importances_[i]

dt_feat_keys = list(dt_feat.keys())
dt_feat_vals = list(dt_feat.values())

sns.barplot(x=dt_feat_keys, y=dt_feat_vals)
plt.title('Feature importance in decision tree classifier')
plt.show()
```

To find the best parameters for the decision tree classifier, we implemented a grid search cross-validation, which is an exhaustive comparison of precision scores with a range of model parameters. In this particular case, the 'best model' came out to be one with `max_depth` of 3, `min_samples_leaf` of 3, and `min_samples_split = 2`. 

```python
params_dict = { 'max_depth':range(30), 'min_samples_split':[1, 2, 3, 4, 5], 'min_samples_leaf':[1, 2, 3, 4, 5]}

grid = GridSearchCV(dt, params_dict, scoring='average_precision')
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.best_estimator_)
```

# V. Business Metrics

Choosing a recipe that is popular can cause as much as a 40% increase in web traffic to other parts of the Tasty Bytes website. 

The current data illustrates a success rate of choosing a popular recipe about 61% of the time, while the two models have a success rate slightly above 80%. Since a recipe is either marked high-traffic or not within a day, a potential KPI is the number of days where a high-traffic recipe is successfully predicted in a fixed period of time. One choice is to count days over a single calendar year. 

In one (365 day) year, we expect the current practice to yield 222.65 days of popular recipe features. By contrast, implementing one of the two models above yield 302.95 and 299.3 popular recipe days for the logistic regression and decision tree models, respectively, which is up to a 36% increase in the number of days projected to have popular recipes in one calendar year. 

```python
sns.barplot(x = ['Current', 'Logistic', 'Decision Tree'], y = [222.65, 302.95, 299.3])
plt.title('Expected number of correct popular recipe predictions (yr.)')
plt.xlabel('Model')
plt.ylabel('Days')
plt.show()
```

# VI. Final Summary & Recommendations
Based on the currently held data, the logistic regression model performed the best with respect to precision, at around 83% for testing data. In turn, this suggests the logistic regression model will perform better with respect to the core KPI of days with high traffic per year. 

We recommend the following:

The logistic regression model places a bit more importance on these variables. If the frequency with which the nutritional information is collected declines, the **decision tree** model would likely perform better, as it places less emphasis on features other than the category.
- Perform an A/B test to gauge the difference in performance from current practice and the proposed model; randomly divide users into two groups, one of which is recommended a recipe from current practice and the other is recommended a recipe predicted to be popular by the model. If there is not a significant difference in performance, the model may need to be re-trained, or more research may need to be done to see if there are other factors driving this surge in daily web traffic.
